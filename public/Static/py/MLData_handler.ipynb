```python
data = pd.read_json('../data/stockinfo3.json')
test_df = pd.DataFrame(data)
test_df
```

```python

count = 378
p=0
for d in data['Ticker'].unique():
    x = data['opening'][p] #pulls opening price list for the currently selected company
    y = data['closing'][p] #pulls closing price list for the currently selected company
    k = 377 #counts through the stock prices in the next for loop
#     j = 378
    for i in range(len(x)):
        try:
            date = x[k][0]
            opening = x[k][1]
            closing = y[k][1]
            ticker = data['Ticker'][p]
            name = data['name'][p]
            sector = data['sector'][p]
            k = k - 1
            next_opening = x[k][1]
#             print(ticker,next_opening)
            if (next_opening == x[-1][1]):
                next_opening = "NaN"
#             else:
#                 pass
            tdf = tdf.append({'Ticker': ticker,
                                'date': date,
                                'sector': sector,
                                'name': name,
                                'opening': opening,
                                'closing': closing,
                                'next_Opening': next_opening}, ignore_index=True)
        
        except:
            pass
    
    p+=1
    

```
